### Colored Scatterplots

```python
iris = pd.read_csv('iris.csv')
iris.sample(5)
```

```python
# relationship between Sepal Length and Petal Length
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

```python
# We have to provide color of every graph point as per its species.
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.7)
plt.xlabel('Sepal Length')# cmap provides different color, c (color) attribute will provide color categorywise
plt.ylabel('Petal Length') # alpha attribute will define opacity of our points
plt.colorbar() # color bar represents the different color value
```

```python
# cmap and alpha
# cmap provides different color, c attribute will divide whole plots in their respective category
# alpha attribute will define opacity of our points
```

### Plot size

```python
plt.figure(figsize=(15,7)) # width, height
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

# Difference in syntax in both ways (plt.scatter & plt.subplots)

```python
plt.figure(figsize=(15,6))
plt.scatter(batters['avg'],batters['strike_rate'])
plt.title('Something')
plt.xlabel('Avg')
plt.ylabel('Strike Rate')
plt.show()
```

```python
# we can make scatter plot using subplots as well
fig,ax = plt.subplots(figsize=(15,6))
ax.scatter(batters['avg'],batters['strike_rate'],color='red',marker='+')
ax.set_title('Something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')
fig.show()
```

### Main power of subplots is that we can plot multiple graphs side by side.

```python
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6)) # 2*1 so we can make 2 graph side by side
# if ncols was 2 then we could make 2*2 = 4 graphs
# We get 2 axis place in a single graph
# sharex=True, this means x-axis labels will not be shown in above graph
# (we will only show xlabels once as both graph contains same values)
ax[0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')


ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))# here we have 4 axis object within the single figure itself
# axis object is a 2-D array so we have to fetch first axis by providing 2 values
ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])  # here we are not sharing x-axis of every graph as every graph has its own x labels
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,sharex=True,figsize=(10,10))
ax
```

```python
# Thrid way of plotting multiple graphs in a single figure
fig = plt.figure()

# we have to make row value as 2 for more than 2 graphs in a single figure
ax1 = fig.add_subplot(2,2,1) # 2nd row, 2nd column, 1st graph
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2) # 2nd row, 2nd column, 2nd graph
ax2.hist(batters['runs'])

ax3 = fig.add_subplot(2,2,3) # 2nd row, 2nd column, 3rd graph
ax3.hist(batters['avg'])
```

### 3D Surface Plots

```python
x = np.linspace(-10,10,100) # 100 points between range -10 to 10
y = np.linspace(-10,10,100) # 100 points between range -10 to 10
```

```python
xx, yy = np.meshgrid(x,y)
print(xx.shape)
print(yy.shape)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis') # cmap is a gradient of colour
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.log(xx)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

### Contour Plots : This plot represents a 3-d graph into a 2-d graph

```python
# We are representing above graph in 2-d
fig = plt.figure(figsize=(12,8))

ax = plt.subplot() # 2-d

p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
# another contour graph way
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p) # blue color shows the bottom and yellow color shows top of 3-d graph
```

```python
# now plotting this 3-d graph in 2-d
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
