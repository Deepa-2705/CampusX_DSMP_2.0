```python
import seaborn as sns
```

```python
# import datasets
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

### 4)  Categorical Plots : In categorical plots also we have 2 ways to plot graph.
- axes level function
- figure level function

### Categorical Scatter Plot (Bivariate)

- Stripplot
- Swarmplot

### Categorical Distribution Plots (Univariate)

- Boxplot
- Violinplot

### Categorical Estimate Plot -> for central tendency

- Barplot
- Pointplot
- Countplot

### Figure level function -> `catplot`

###

```python
# Generally we use scatter plot on numerical dataset
sns.scatterplot(data=tips, x='total_bill',y='tip')
```

```python
# strip plot : We can plot a scatter plot with categorical columns (Condition: It should contains one numerical and one categorical column)
# axes level function
sns.stripplot(data=tips,x='day',y='total_bill')
```

### We can see w.r.t x- axis our data points are spreading but it should be place in a single line.
### This is because seaborn adds some random noise internally so that it can be more readable.
### To correct this we use jitter than all data point will arrange in a single line.

```python
sns.stripplot(data=tips,x='day',y='total_bill',jitter=False)
```

```python
# using catplot
# figure level function
sns.catplot(data=tips, x='day',y='total_bill',kind='strip')
```

```python
# jitter
sns.catplot(data=tips, x='day',y='total_bill',kind='strip',jitter=0.2,hue='sex') # By increasing jitter value from 0 it will spread the data points more
```

```python
# swarmplot : It internally run some algorithms which makes visualization to show the distribution of the dataset
# figure level function
sns.catplot(data=tips, x='day',y='total_bill',kind='swarm')
```

```python
# hue
# axis level function
sns.swarmplot(data=tips, x='day',y='total_bill',hue='sex')
```

### Swarmplot is suitable for smaller dataset. But stripplot works suitably for medium and larger dataset.

### Boxplot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

image.png

```python
# Box plot : It usually tells the distribution of single column only but in seaborn we have availability to plot it with another categorical column.
# axis level function
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
# Using catplot
# figure level function
sns.catplot(data=tips,x='day',y='total_bill',kind='box')
```

```python
# hue
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
# single boxplot -> numerical col
sns.boxplot(data=tips,y='total_bill')
```

### Violinplot = (Boxplot + KDEplot)

```python
# violinplot
# axes level fucntion
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
# figure level function
sns.catplot(data=tips,x='day',y='total_bill',kind='violin') # symmetrical distribution both side of boxplot(center)
```

```python
# hue
# side by side 2 different distribution based on sex column of dataset
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex',split=True)
```

```python
# barplot : We can calculate some aggregated value of one categorical column w.r.t one numerical column
# some issue with errorbar
import numpy as np
sns.barplot(data=tips, x='sex', y='total_bill',hue='smoker',estimator=np.std) # on y axis : standard deviation of total bill
```

### The black line is error bar. This represents that this categories repeats in our dataset in a single column for every totalbill.
### We can remove this errorbar by using parameter ci=None

```python
# point plot : We are connecting the similar bars of barplot with a line then it becomes pointplot
sns.pointplot(data=tips, x='sex', y='total_bill',hue='smoker',estimator=np.std)
```

```python
sns.barplot(data=tips, x='sex', y='total_bill',ci=None)
```

```python
sns.pointplot(data=tips, x='sex', y='total_bill',estimator=np.std)
```

```python
sns.pointplot(data=tips, x='sex', y='total_bill',estimator=np.std,ci=None)
```

### We can remove this errorbar by using parameter ci=None

When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars

```python
# countplot
sns.countplot(data=tips,x='sex',hue='day')
```

A special case for the bar plot is when you want to show the number of observations in each category rather than computing a statistic for a second variable. This is similar to a histogram over a categorical, rather than quantitative, variable

```python
# faceting using catplot
# boxplot for male and female separately
sns.catplot(data=tips, x='sex',y='total_bill',col='smoker',kind='box',row='time')
```

### 5) Regression Plots

- regplot
- lmplot

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval (error scope) for that regression.

```python
# axes level
# hue parameter is not available here
sns.regplot(data=tips,x='total_bill',y='tip')
```

```python
# figure level function
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex') # We can plot separate regression line for every category of column that passed in hue parameter
```

```python
# residplot : This will tells about the error quantity made by datapoints w.r.t regression line
# all data points above central line having more value than actual and all data points below central line having less value than actual
# axes level function
sns.residplot(data=tips,x='total_bill',y='tip')
```

### A second way to plot Facet plots -> FacetGrid

```python
# Facet grid
g = sns.FacetGrid(data=tips,col='day',row='time',hue='smoker')
g.map(sns.violinplot,'sex','total_bill') # g is a facet grid object
```

```python
g = sns.FacetGrid(data=tips,col='day',row='time',hue='smoker')
g.map(sns.scatterplot,'sex','total_bill') # g is a facet grid object
g.add_legend()
```

```python
# Facet grid
g = sns.FacetGrid(data=tips,col='day',row='time')
g.map(sns.boxplot,'sex','total_bill') # g is a facet grid object
```

### Plotting Pairwise Relationship (PairGrid Vs Pairplot)

```python
# pairplot : It will first detect all numerical columns in our dataset and the plot scatter plot pair by pair on every numerical columns
sns.pairplot(iris)
```

###  Graph above diagonal and Graph below diagonal are symmetrical

```python
sns.pairplot(iris,hue='species') #  only histogram or kde plot for diagonal elements
 # it can plot any kind of graph for off diagonal elements
```

### Pairplot works because of pairgrid. There are more features in pairgrid as compare to pairplot.

```python
# pair grid
g = sns.PairGrid(data=iris,hue='species')
# g.map
g.map(sns.scatterplot)
```

```python
# map_diag -> map_offdiag
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.boxplot) # histogram for diagonal elements (it can plot any kind of graph)
g.map_offdiag(sns.kdeplot) # scatter plot for non diagonal elements (it can plot any kind of graph)
```

```python
# map_diag -> map_upper -> map_lower
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot) # histogram plot for non diagonal elements (it can plot any kind of graph)
g.map_upper(sns.kdeplot) # 2-d kde plot for above diagonal elements (it can plot any kind of graph)
g.map_lower(sns.scatterplot) # scatter plot for below diagonal elements (it can plot any kind of graph)
```

```python
# vars
g = sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width']) # we can restrict plotting to our desired columns only
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

### JointGrid Vs Jointplot

```python
# jointplot : It will plot scatter plot for both numerical columns and along with this it will show histogram for both columns separately
sns.jointplot(data=tips,x='total_bill',y='tip',kind='hist',hue='sex')
```

```python
# jointplot : It will plot scatter plot for both numerical columns and along with this it will show histogram for both columns separately
sns.jointplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex')
```

```python
# jointplot : It will plot scatter plot for both numerical columns and along with this it will show histogram for both columns separately
sns.jointplot(data=tips,x='total_bill',y='tip',kind='reg',hue='sex') # regplot was earlier supported not nowonwards
```

```python
# jointgrid
g = sns.JointGrid(data=tips,x='total_bill',y='tip')
g.plot(sns.kdeplot,sns.violinplot) # we can provide any plot here
```
