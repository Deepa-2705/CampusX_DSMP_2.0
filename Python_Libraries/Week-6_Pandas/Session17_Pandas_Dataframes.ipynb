### Selecting rows from a DataFrame

- **iloc** - searches using index positions
- **loc** - searches using index labels

```python
# single row
movies.iloc[5] # 6th row of data (series)
```

```python
# multiple row
movies.iloc[5:15:2]
```

```python
# fancy indexing
movies.iloc[[0,4,5]]
```

```python
# loc
students
```

```python
students.loc['nitish']
```

```python
students.loc['nitish':'rishabh':2] # rishab is included
```

## Difference between loc and iloc:
On giving a range, in iloc last item is excluded but in loc last item is included.

```python
# fancy indexing
students.loc[['nitish','ankita','rupesh']]
```

```python
# fancy indexing using iloc on this student data
students.iloc[[0,3,4]] # index 0 and nitish is pointing to same index internally.
```

### Filtering a DataFrame

```python
ipl.head(2)
```

```python
# find all the final winners
# mask = ipl['MatchNumber'] == 'Final'
# new_df = ipl[mask]
# new_df[['Season','WinningTeam']]

ipl[ipl['MatchNumber'] == 'Final'][['Season','WinningTeam']]
```

```python
# how many super over finishes have occured
ipl[ipl['SuperOver'] == 'Y'].shape[0]
```

```python
# how many matches has csk won in kolkata
ipl[(ipl['City'] == 'Kolkata') & (ipl['WinningTeam'] == 'Chennai Super Kings')].shape[0]
```

```python
# toss winner is match winner in percentage
(ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]/ipl.shape[0])*100
```

```python
# movies with rating higher than 8 and votes>10000
movies[(movies['imdb_rating'] > 8) & (movies['imdb_votes'] > 10000)].shape[0]
```

```python
# Action movies with rating higher than 7.5
# mask1 = movies['genres'].str.split('|').apply(lambda x:'Action' in x) # converting series into string to apply split
mask1 = movies['genres'].str.contains('Action') # converting series into string to apply contains function
mask2 = movies['imdb_rating'] > 7.5

movies[mask1 & mask2]
```
