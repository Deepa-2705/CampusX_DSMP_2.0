# Data Gathering using API

```python
import pandas as pd
import requests
```

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1')
```

```python
# Explicitly converting the incoming data into json format
response.json()
```

```python
response.json()['results'] # extract only result list
```

# Convert this json format into pandas dataframe
image.png

## with these columns and total shape will be 8851 x 7

```python
# convert this result list into dataframe (extract only desired columns)
temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
```

```python
temp_df
```

```python
# create a empty dataframe
df = pd.DataFrame()
df
```

```python
# Go to every web page (total 428 pages)
for i in range(1,429):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df = df._append(temp_df,ignore_index=True) # ignoring index because everytime it will print the same index value (1-20)
```

```python
df
```

```python
df.to_csv('movies.csv')
```
